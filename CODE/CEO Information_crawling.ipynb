```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import re
import time 
import datetime 
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
path=r'C:/Users/NTX550/chromedriver-win64/chromedriver.exe'
s=Service(path)
driver=webdriver.Chrome(service=s)
driver.set_window_size(700,700)
```

# 링크드인

```python
df=pd.read_excel('C:/Users/NTX550/Desktop/linkedin.xlsx')
df['linkein_list']= ''

for idx in range(1154,1200):
    url=df.at[idx,'link']
    driver.get(url)
    
    time.sleep(5)
    soup=BeautifulSoup(driver.page_source)
    
    #링크드인
    if 'linkedin' in url:
        school=soup.find_all("span", class_="visually-hidden")
        sli=[]
        for i in range(len(school)):
            
            if school[i].text=='학력':
                for j in range(i+1,i+3):
                    sli.append(school[j].text)
                break
        df.at[idx,'linkein_list']=sli
        print(sli)
        sli=[]
    print("-"*50)
print("done!!")
```

```python
df.to_excel('linkedin.xlsx')
```

# 로켓펀치

```python
df=pd.read_excel('C:/Users/NTX550/Desktop/rocketpunch.xlsx')
url = 'https://www.rocketpunch.com/@1322dbff6f9041a1'
driver.get(url)
```

```python
df['rocketpunch_list']= ''

for idx in range(2,10):
    url=df.at[idx,'link']
    driver.get(url)
    
    time.sleep(3)
    soup=BeautifulSoup(driver.page_source)
    
    if 'rocketpunch' in url:
        rocket=soup.find('div',class_='ui divided segment education items').find_all('div',class_='school-name title')
        rli=[]
        for i in range(len(rocket)):
            rli = rocket[i].find('strong').text
            
            if rocket[i].text=='대학':
                s=soup.find('div',class_='ui divided segment education items').find_all('div',class_='break desc')
                for i in range(len(s)):
                    rli.append(s[i].text.replace('\n','').rstrip().lstrip())

        print(rli)
        rli=[]

    print("-"*50)
print("done!!")
```

```python
df.to_excel('rocketpunch.xlsx')
```
